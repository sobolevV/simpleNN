```python

class NN:
    def __init__(self, inputs, layers_sizes, outputs, learning_rate=0.1):
        self.inputs = inputs
        self.layers_sizes = layers_sizes
        self.layers_count = len(layers_sizes)+1
        self.outputs = outputs
        self.lr = learning_rate
        
        # Empty layers
        # self.calc_layers = None
        
        # INIT WEIGHTS. left weights
        self.weights = [np.random.rand(layers_sizes[0], inputs)]
        # middle weights
        for i in range(self.layers_count-2):
            self.weights.append( np.random.rand(self.layers_sizes[i+1], self.layers_sizes[i]))
        self.weights.append( np.random.rand(self.outputs, self.layers_sizes[-1])) 
        
        # INIT BIASES
        self.biases = []
        for i in range(self.layers_count-1):
            self.biases.append( np.random.rand(self.layers_sizes[i], 1) )
        self.biases.append( np.random.rand(self.outputs, 1) )    
    
    def train(self, iter_data, num_iterations):
        model_size = len(layers_sizes)+1
        for ep in range(num_iterations):
            x_train, y_index = next(iter_data)
            
            x_train = np.reshape(x_train, (self.inputs, 1))
            x_train = (x_train.astype(float) / 127.5) - 1.
            
            y_train = np.zeros((self.outputs, 1), dtype=float) 
            y_train[y_index] = 1.0
            
            out = self.feedforward(x_train)
            # print(np.array([out, y_train]))
            # Ошибка
            error = y_train - out
            if ep % 100 == 0:
                print(f"ep: {ep}, err: {np.sum(error**2)/error.shape[0]}") # /error.shape[0]
            
            self.backpropagation(error)
            
    
    def sigmoid(self, layer):
        s = 1 / (1 + np.exp(-1*layer))
        return s

    # delta sigmoid (sigma')
    def sigmoid_deriv(self, layer):
        return layer * (1 - layer)

    # gradient for layer (error and calculated value)
    def gradient(self, err, y_res):
        gradient_layer = err*self.sigmoid_diff(y_res)
        return gradient_layer 
    
    def feedforward(self, x_input):
        self.calc_layers = [x_input]
        for i in range(self.layers_count):
            layer = self.sigmoid(np.dot(self.weights[i], self.calc_layers[i]) )  #  + self.biases[i]
            self.calc_layers.append(layer)
        return self.calc_layers[-1]

    def backpropagation(self, error):
        # Обновление весов и байесов __ LAYERS + WEIGHTS
        # цикл с конца до начала по кол-ву слоев
        for step in range(1, self.layers_count+1):

            # градиент на i слое
            layer_gradient = -error * (self.calc_layers[-step] * (1 - self.calc_layers[-step]))
            
            # delta_weights 
            delta_w = np.dot(layer_gradient, self.calc_layers[-step-1].T)
            self.weights[-step] += self.lr*delta_w
            
            # ошибка на скрытом слое
            error = np.dot(self.weights[-step].T, error)
            
            # обновляем байес. связи
            # self.biases[-step] -= self.lr*layer_gradient
```

```python
def sigmoid(layer):
    # res layer summ (first red)
    # output_sum = np.sum(layer)
    return 1 / (1 + np.exp(-layer))

    # delta sigmoid (second red)
def sigmoid_diff(layer):
    res_sigm = sigmoid(layer)
    return res_sigm * (1 - res_sigm)

    # -(ej)
def error(y_true, y_res):
    return y_true - y_res

    # formuld (all in one)
def gradient(err, y_res):
    # gradient layer #, prev_layer
    gradient_layer = err*sigmoid_diff(y_res)
    return gradient_layer 
```
